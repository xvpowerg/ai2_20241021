# Series
Series與NumPy陣列非常相似（事實上它是建立在NumPy陣列對象之上的）。將NumPy陣列與Series區分開來的是，Series可以具有軸標籤，這意味著它可以通過標籤進行索引，而不僅僅是通過數字位置。它也不需要保存數字數據，可以保存任意的Python對象。


```python
import pandas as pd
```

```python
myIndex = ['USA','Canada','Mexico']
myData = [1776,1867,1821]
myser = pd.Series(data=myData)
print(myser)
```

```python
myser = pd.Series(data=myData,index=myIndex)
print(myser)
```

```python
import numpy as np
ran_data = np.random.randint(0,100,4)
ran_data
```

```python
names = ["Ken","Vivin","Lucy","Joy"]
ages = pd.Series(ran_data,names)
ages
```

```python
ags = {"Gigi":5,"Any":10,"Carlos":15}
ageSer = pd.Series(ags)
ageSer
```

```python
q1 = {"Japan":80,"China":450,"India":200,"USA":250}
q2 = {"Brazil":100,"China":500,"India":210,"USA":260}
sales_Q1 = pd.Series(q1)
sales_Q2 = pd.Series(q2)
print(sales_Q1["Japan"])
print(sales_Q1[0])
print(sales_Q2['Brazil'])
#print(sales_Q1['Brazil'])
```

```python
sales_Q2.keys()
```

```python
sales_Q2 * 2
```

```python
sales_Q1 + sales_Q2
```

```python
sales_Q1.add(sales_Q2,fill_value=0)
```

# Conditional Filtering

```python
import pandas as pd
import numpy as np
df = pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
bool_series = df["total_bill"] > 30
df[bool_series].head()
```

```python
df[df["total_bill"] > 30].head()
```

```python
df[df['sex'] == "Male"].head()
```

```python
df[(df['total_bill'] > 30) & (df['sex'] == 'Male')]
```

```python
df[(df['day'] == 'Sun') | (df['day'] == 'Sat')]
```

```python
options = ['Sat','Sun']
df[df['day'].isin(options)]
```
